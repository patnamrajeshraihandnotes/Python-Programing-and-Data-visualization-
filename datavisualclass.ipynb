# Bar Graph 

```python
from matplotlib import pyplot as plt
import numpy as np
product={'mobiles':50,'cars':60,'bus':100,'playtoys':100}
name=list(product.keys())
quantity=list(product.values())
plt.bar(name,quantity,color='orange')
plt.title('Kids Toys')
plt.xlabel('product')
plt.ylabel('quantity')

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
labels=['g1','g2','g3','g4','g5']
mens_means=[20,35,30,35,27]
women_means=[25,32,34,20,27]
mens_std = [2,3,4,1,2]
women_std =[3,5,2,3,3]
width =0.1      
fig,ax = plt.subplots()
ax.bar(labels, mens_means, width, yerr=mens_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=mens_means,label='Women')
ax.set_ylabel('scores')
ax.set_title('scores by group and gender')
ax.legend()
plt.show()
```

# Box plot

```python
from matplotlib import pyplot as plt
import numpy as np
one = [1,2,3,4,5,6,7,8,9]
two = [2,5,6,4,7,8,9,1,2]
three = [3,2,5,4,6,7,9,1,2]
data =([one,two,three])
plt.boxplot(data)
plt.show()
```
