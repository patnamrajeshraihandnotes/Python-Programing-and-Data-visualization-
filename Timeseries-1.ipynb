```python
import pandas as pd
df=pd.read_csv("E:/aapl1.csv",parse_dates=["Date"],index_col="Date")
df.head(5)
```

```python
df.index
```

# To Retrive Jan 2017 Stock Prices

```python
df["2017-01"]
```

```python
df.index
```

```python
df["2017-01-03"] # particular date index
```

```python
df["2017-01-03":"2017-01-10"] # particular date range index
```

```python
# Monthly stock prizes

df.resample('M').mean()
```

```python
# Monthly stock prizes to plot chart

df.resample('M').mean().plot()
```

```python
# # Weekly stock prizes

df.resample('W').mean()
```

```python
# # Weekly stock prizes

df.resample('W').mean().plot()
```

```python
# # Quarterly stock prizes

df.resample('Q').mean().plot()
```

```python
# # Quarterly stock prizes change the plot in to bar

df.resample('Q').mean().plot(kind="bar")
```
